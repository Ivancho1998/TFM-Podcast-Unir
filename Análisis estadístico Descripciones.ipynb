```python
# prompt: en una columna tengo en esquema de  temas que estan numerdos asi 1. 2. 3. 4. 5. quiero sacar cada uno indepnedinete de cada fila son 5

import pandas as pd

# Assuming your DataFrame is named 'df' and the column with numbered items is named 'Temas'
# Replace 'Temas' with the actual column name if it's different

def extract_topics(df):
    """
    Extracts individual topics from a column containing numbered items.

    Args:
        df: Pandas DataFrame with a 'Temas' column.

    Returns:
        A new DataFrame with each topic in a separate column.
    """
    # Handle missing values by creating an empty string for these columns
    df['temas'] = df['temas'].fillna('')

    # Extract topics and add them to new columns
    for i in range(1, 6):  # Assuming there are 5 topics
      df[f'tema_{i}'] = df['temas'].str.extract(f'{i}\\.\s*(.*?)(?=\s*{i+1}\\.|$)', expand=False).fillna('')

    return df


# Example usage assuming your DataFrame is named df
# Replace this with the actual column name if it's different from 'Temas'

#if 'Temas' in df.columns: #Check if the 'Temas' column exists
df = extract_topics(df)
df
#else:
#  print("Column 'Temas' not found in DataFrame.")
```

```python
# prompt: quiero preprocesar la data de manera que cada palabra sea trabajada individualmente y eliminando conecotres solo dejando lo de valor para despues  generar un grafico de esos que se muestran las palabras y que son mas grandes cuando aparecen mas

import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import re

# ... (your existing code for importing data from MongoDB) ...

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation and numbers
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    # Remove stop words (you might need to customize this list)
    stop_words = ['el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'de', 'del', 'al', 'en', 'y', 'o', 'a', 'se', 'es', 'que', 'con', 'por', 'para', 'sin', 'sobre', 'entre', 'durante', 'desde', 'hasta', 'contra', 'seg√∫n', 'como', 'pero', 'aunque', 'si', 'no', 'ni', 'o', 'u', 'mas', 'menos']
    words = [word for word in text.split() if word not in stop_words]

    return " ".join(words)


# Apply preprocessing to relevant columns (e.g., 'tema_1' to 'tema_5')

for i in range(1, 6):
    column_name = f'tema_{i}'
    if column_name in df.columns:
        df[column_name] = df[column_name].astype(str).apply(preprocess_text)  # Convert to string before preprocessing
    else:
        print(f"Warning: Column '{column_name}' not found in DataFrame.")

# Combine all preprocessed text data
all_text = " ".join(str(value) for i in range(1, 6) if f'tema_{i}' in df.columns for value in df[f'tema_{i}'].astype(str).values)

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Display the generated image:
plt.figure(figsize=(10, 5), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# prompt: ahora de cada columna creada de temas quiero generar un grafico de esos que se muestran las palabras y que son mas grandes cuando aparecen mas

import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pandas as pd

# ... (your existing code to load the data into the DataFrame 'df') ...

def generate_wordclouds(df):
    """Generates word clouds for each topic column in the DataFrame."""

    for i in range(1, 6):  # Assuming up to 5 topics
        column_name = f'tema_{i}'
        if column_name in df.columns:
            text = ' '.join(df[column_name].astype(str))  # Convert to string and concatenate
            if text.strip():  # Check if the text is not empty
              wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
              plt.figure(figsize=(10, 5))
              plt.imshow(wordcloud, interpolation='bilinear')
              plt.axis("off")
              plt.title(f'Word Cloud for Tema {i}')
              plt.show()
            else:
              print(f"No data for Tema {i}. Skipping wordcloud generation.")
        else:
            print(f"Column '{column_name}' not found in DataFrame.")


# Example usage
generate_wordclouds(df)
```

```python
# prompt: tengo una variable que se llama rangos de edad que estas asi numero-numero y quiero graficarla en un diagrama de barras , las que nos son de ese estilo numero-numero, dejalas como no se puede determinar  pero que aparezca en la grafica asi

# ... (your existing code) ...

import matplotlib.pyplot as plt

def plot_age_ranges(df):
    """Plots age ranges as a bar chart, handling invalid formats."""

    def categorize_age(age_range):
        if pd.isna(age_range):
          return "No se puede determinar"
        if re.match(r'\d+-\d+', str(age_range)):
            return age_range
        else:
            return "No se puede determinar"

    if 'rango_edad' in df.columns:
        df['categorized_age'] = df['rango_edad'].apply(categorize_age)
        age_counts = df['categorized_age'].value_counts()

        plt.figure(figsize=(12, 6))
        age_counts.plot(kind='bar')
        plt.xlabel('Rangos de Edad')
        plt.ylabel('Frecuencia')
        plt.title('Distribuci√≥n de Rangos de Edad')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    else:
        print("Column 'rangos de edad' not found in the DataFrame.")

# Example usage
plot_age_ranges(df)

# ... (rest of your code) ...
```
