```python
# Lista de títulos en español
titulos = [
    "Impacto del cambio climático en la biodiversidad",
    "Avances en la inteligencia artificial en el sector salud",
    "Estrategias para mejorar la educación en zonas rurales",
    "Análisis del crecimiento económico en América Latina"
]

# Función para extraer temas y sentimiento de un título
def extraer_temas_y_sentimiento(titulo):
    messages = [
        {"role": "system", "content": (
            "Tu eres un creador de contenido experto que también conoces sobre temas de actualidad. "
            "Estás entrenado para analizar el sentimiento (positivo, negativo o neutral) del título. "
            "Además, analizas los 5 temas principales relacionados con el título y los devuelves como respuesta."
            "Tu objectivo es ser lo mas viral posible tus videos y por eso lo analizas los temas"
        )},
        {"role": "user", "content": f"Quiero una lista de los 5 temas principales que estén relacionados con el título y además me analices qué tipo de sentimiento atrae el título. El formato debe ser: 'temas: [lista de temas], sentimiento: [sentimiento]'. Título: '{titulo}'."}
    ]

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        max_tokens=100,
        temperature=0.3
    )

    return response.choices[0].message.content.strip()

# Crear un DataFrame con los resultados
data = []

for titulo in titulos:
    resultado = extraer_temas_y_sentimiento(titulo)
    temas, sentimiento = resultado.split(", sentimiento: ")
    temas = temas.replace("temas: ", "").strip("[]")  # Limpiar la lista de temas
    data.append({"Título": titulo, "Temas": temas, "Sentimiento": sentimiento.strip()})

df = pd.DataFrame(data)

```

```python
# Lista de títulos en español
titulos = [
    "Impacto del cambio climático en la biodiversidad",
    "Avances en la inteligencia artificial en el sector salud",
    "Estrategias para mejorar la educación en zonas rurales",
    "Análisis del crecimiento económico en América Latina"
]

# Función para determinar si un título es sensacionalista
def evaluar_sensacionalismo(titulo):
    messages = [
        {"role": "system", "content": (
            "Eres un experto en creación de contenido y análisis de medios. "
            "Tu función principal es identificar si un título utiliza medios sensacionalistas para llamar la atención. "
            "Tu objetivo es evaluar cómo el título puede atraer viralidad en redes sociales o medios."
        )},
        {"role": "user", "content": (
            f"Quiero que determines si este título está utilizando medios sensacionalistas para llamar la atención. "
            f"El formato de tu respuesta debe ser: 'sensacionalismo: [sí/no/no se puede inferir]'. "
            f"Título: '{titulo}'."
        )}
    ]

    # Llamada a la API
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        max_tokens=50,
        temperature=0.3
    )

    # Acceder al contenido del mensaje generado por la API
    return response.choices[0].message.content.strip()

# Crear un DataFrame con los resultados
data = []

for titulo in titulos:
    try:
        resultado = evaluar_sensacionalismo(titulo)
        # Extraer el valor de "sensacionalismo" de la respuesta
        if "sensacionalismo:" in resultado:
            sensacionalismo = resultado.split("sensacionalismo:")[1].strip()
        else:
            sensacionalismo = "Error en la respuesta"
    except Exception as e:
        sensacionalismo = f"Error: {str(e)}"

    # Agregar los resultados al dataset
    data.append({"Título": titulo, "Sensacionalismo": sensacionalismo})

# Convertir los datos a un DataFrame
df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)
```

```python
# Función genérica para realizar llamadas a la API
def analizar_titulo(titulo, prompt_system, prompt_user):
    messages = [
        {"role": "system", "content": prompt_system},
        {"role": "user", "content": prompt_user.format(titulo=titulo)}
    ]

     # Llamada a la API
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        max_tokens=50,
        temperature=0.3
    )

    # Acceder al contenido del mensaje generado por la API
    return response.choices[0].message.content.strip()

# Función para extraer los temas principales
def extraer_temas(titulo):
    prompt_system = (
        "Eres un creador de contenido experto que también conoces sobre temas de actualidad. "
        "Estás entrenado para analizar los 5 temas principales relacionados con el título. "
        "Tu objetivo es ser lo más viral posible con tus videos y por eso analizas los temas."
    )
    prompt_user = (
        "Quiero una lista de los 5 temas principales relacionados con el título. "
        "El formato debe ser: 'temas: [lista de temas]'. "
        f"Título: '{titulo}'."
    )
    return analizar_titulo(titulo, prompt_system, prompt_user)

# Función para analizar el sentimiento
def extraer_sentimiento(titulo):
    prompt_system = (
        "Eres un creador de contenido experto y estás entrenado para analizar el sentimiento (positivo, negativo o neutral) del título. "
        "Tu objetivo es evaluar el sentimiento que atrae el título para maximizar su viralidad."
    )
    prompt_user = (
        "Analiza el sentimiento que transmite el título. "
        "El formato de la respuesta debe ser: 'sentimiento: [sentimiento]'. "
        f"Título: '{titulo}'."
    )
    return analizar_titulo(titulo, prompt_system, prompt_user)

def evaluar_sensacionalismo(titulo):
    prompt_system = (
        "Eres un experto en creación de contenido y análisis de medios. "
        "Tu función principal es identificar si un título utiliza medios sensacionalistas para llamar la atención. "
    )
    prompt_user = (
        "Quiero que determines si este título está utilizando medios sensacionalistas para llamar la atención. "
        "El formato de tu respuesta debe ser: 'sí/no/no se puede inferir'. "
        "Título: '{titulo}'."
    )
    return analizar_titulo(titulo, prompt_system, prompt_user)

def evaluar_palabras_negativas(titulo):
    prompt_system = (
        "Eres un experto en la creación de contenido y tu función principal es determinar "
        "si este título está utilizando la estrategia de usar palabras negativas para llamar la atención. "
    )
    prompt_user = (
        "Título: '{titulo}'. "
        "¿Está utilizando palabras negativas para llamar la atención? Responde con 'sí', 'no' o 'no se puede inferir'."
    )
    return analizar_titulo(titulo, prompt_system, prompt_user)

# Lista de emociones posibles
emociones_posibles = [
    'esperanza', 'temor', 'curiosidad', 'alegría', 'tristeza', 'ira', 'miedo', 'sorpresa',
    'confusión', 'asco', 'frustración', 'interés', 'ansiedad', 'entusiasmo', 'desesperación',
    'satisfacción', 'fascinación', 'desconfianza', 'melancolía', 'optimismo', 'desilusión',
    'orgullo', 'vergüenza', 'culpa', 'compasión', 'resentimiento'
]

def evaluar_emociones(titulo):
    prompt_system = (
            "Eres un experto en análisis de emociones en contenido. "
            "Tu tarea es identificar la emoción principal que transmite este título. "
            "Tu respuesta debe ser una de las siguientes emociones: "
            f"{', '.join(emociones_posibles)}."
    )
    prompt_user = (
        "Título: '{titulo}'. "
        "¿Qué emoción principal transmite este título? y quiero que la respuesta sea solo la emocion nada mas"
    )
    return analizar_titulo(titulo, prompt_system, prompt_user)

def evaluar_rango_edad(titulo):
    prompt_system = (
        "Eres un experto en marketing y análisis de audiencias. "
        "Tu tarea es identificar a qué rango de edad parece estar dirigido este título. "
        "Tu respuesta debe ser entre estos rangos '13-19', '20-35', '36-45', '46-65'."
    )
    prompt_user = (
        "Título: '{titulo}'. "
        "¿A qué rango de edad parece estar dirigido este título? Responde con uno de los siguientes rangos: "
        "'13-19', '20-35', '36-45', '46-65'. solo el rango nada de texto"
    )
    return analizar_titulo(titulo, prompt_system, prompt_user)

# Ajustar el procesamiento del DataFrame
def procesar_titulos(df):
    resultados = []
    for _, row in df.iterrows():
        titulo = row["title"]
        print(f"Procesando título: {titulo}")
        # Inicializar valores predeterminados
        temas = "No identificado"
        sentimiento = "No identificado"
        sensacionalismo = "No identificado"
        palabras_negativas = "No identificado"
        emocion = "No identificado"
        rango_edad = "No identificado"

        try:
            # Extraer temas
            temas_response = extraer_temas(titulo)
            if "temas:" in temas_response:
                temas = temas_response.replace("temas: ", "").strip("[]")
            else:
                print(f"Formato inesperado en temas: {temas_response}")
        except Exception as e:
            print(f"Error al extraer temas para el título '{titulo}': {e}")

        try:
            # Extraer sentimiento
            sentimiento_response = extraer_sentimiento(titulo)
            if "sentimiento:" in sentimiento_response:
                sentimiento = sentimiento_response.replace("sentimiento: ", "").strip()
            else:
                print(f"Formato inesperado en sentimiento: {sentimiento_response}")
        except Exception as e:
            print(f"Error al extraer sentimiento para el título '{titulo}': {e}")

        try:
            # Evaluar sensacionalismo
            sensacionalismo = evaluar_sensacionalismo(titulo)
        except Exception as e:
            print(f"Error al procesar sensacionalismo para el título '{titulo}': {e}")

        try:
            # Evaluar palabras negativas
            palabras_negativas = evaluar_palabras_negativas(titulo)
        except Exception as e:
            print(f"Error al procesar palabras negativas para el título '{titulo}': {e}")

        try:
            # Evaluar emoción
            emocion = evaluar_emociones(titulo)
        except Exception as e:
            print(f"Error al procesar emoción para el título '{titulo}': {e}")

        try:
            # Evaluar rango de edad
            rango_edad = evaluar_rango_edad(titulo)
        except Exception as e:
            print(f"Error al procesar rango de edad para el título '{titulo}': {e}")

        # Agregar resultados
        resultados.append({
            "video_id": row["video_id"],
            "title": titulo,
            "temas_openai": temas,
            "sentimiento_openai": sentimiento,
            "sensacionalismo_openai": sensacionalismo,
            "palabras_negativas_openai": palabras_negativas,
            "emocion_openai": emocion,
            "rango_edad_openai": rango_edad
        })

    # Crear un DataFrame con los resultados
    return pd.DataFrame(resultados)
# Aplicar el procesamiento
#df_resultados = procesar_titulos(first_five_rows)

```
